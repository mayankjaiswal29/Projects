## **Data Cleaning**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://github.com/mayankjaiswal29/Datasets/raw/main/AB_NYC_2019.csv')
```

```python
df = df.fillna('NA')
```

```python
import seaborn as sns
sns.boxplot(x =df['minimum_nights'])
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(6, 4))
ax.scatter(df['price'], df['minimum_nights'])
plt.show()
```

```python
outlier_indices = np.where((df['price'] >= 6000 ) & (df['minimum_nights'] >= 600))

no_outliers = df.drop(outlier_indices[0])

# Scatter plot without outliers
fig, ax_no_outliers = plt.subplots(figsize=(6, 4))
ax_no_outliers.scatter(no_outliers['price'], no_outliers['minimum_nights'])
plt.show()
```
